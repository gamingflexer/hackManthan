### Creating Single Function for Finding Rank

```python
def CalculateRank(lat, long, recent_days):
    recent_time =  date.fromisoformat('2021-06-30')-timedelta(days=recent_days)
    F = len(df.loc[(df['Latitude'] == lat) & (df['Longitude'] == long)])
    R = len(df.loc[(df['Latitude'] == lat) & (df['Longitude'] == long) & (df['date'] >= recent_time)])
    N = len(df.loc[(df['Latitude'] > lat-0.0011) & (df['Latitude'] < lat+0.0011) 
               & (df['Longitude'] > long-0.0011) & (df['Longitude'] < long+0.0011)])
    Fw = 0.2
    Rw = 0.7
    Nw = 0.2
    return (F*Fw) + (R*Rw)+ (N*Nw)
```
