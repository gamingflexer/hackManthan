```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/cosmos/Desktop/HackManthan/python/data.csv")
```

```python
df.columns
```

```python
df=df.drop(axis=0,columns=["Unnamed: 0","Event"])
df.head(3)
```

```python
df['Create Date/Time'] = pd.to_datetime(df['Create Date/Time'], format="%Y-%m-%d %H:%M:%S")

df['year']  = df['Create Date/Time'].dt.year
df['month'] = df['Create Date/Time'].dt.month
df['day']   = df['Create Date/Time'].dt.day
df['hour'] = df['Create Date/Time'].dt.hour
df['minute'] = df['Create Date/Time'].dt.minute
df['second'] = df['Create Date/Time'].dt.second
df['day_of_week']  = df['Create Date/Time'].dt.dayofweek
df['date'] = df['Create Date/Time'].dt.date

dw_mapping={
    0: 'Monday', 
    1: 'Tuesday', 
    2: 'Wednesday', 
    3: 'Thursday', 
    4: 'Friday',
    5: 'Saturday', 
    6: 'Sunday'
} 
df['day_of_week_name'] = df['day_of_week'].map(dw_mapping)
df.head()
```

```python
df=df.drop(axis=0,columns=["Create Date/Time"])
df=df.drop(axis=0,columns=["date"])
df=df.drop(axis=0,columns=["day_of_week_name"])
df=df.drop(axis=0,columns=["Latitude","Longitude"])
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['Circle'] = le.fit_transform(df['Circle'])
df['Police Station'] = le.fit_transform(df['Police Station'])
df['Caller Source'] = le.fit_transform(df['Caller Source'])
df['Event Type'] = le.fit_transform(df['Event Type'])
df['Event Sub-Type'] = le.fit_transform(df['Event Sub-Type'])
df['District'] = le.fit_transform(df['District'])

df.head()
```

```python
df
```

```python
df['violent_crime_occurence'] = 0
```

```python
violent_crimes = list(map(float, df['Event Type']))
violent_crimes_mean = sum(violent_crimes)/len(violent_crimes)
violent_crimes_mean
```

```python
df['mean_violent_crimes'] = violent_crimes_mean
df['violent_crime_occurence'] = np.where(violent_crimes>=df['mean_violent_crimes'], '1', '0')
df.groupby('violent_crime_occurence').mean()
```
